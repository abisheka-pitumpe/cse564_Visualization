```python
import pandas as pd
from fuzzywuzzy import process

# Load the existing DataFrame
# existing_df = pd.read_csv("existing_data.csv")

# Load the population data from the CSV file
population_df = pd.read_csv("population2024.csv")
# Filter existing DataFrame for the year 2024
df_2024 = df[df['Year'] == 2024]


# Function for fuzzy matching
def fuzzy_match_country(country_name, choices):
    match = process.extractOne(country_name, choices)
    return match[0] if match[1] >= 80 else None  # Adjust threshold as needed

# Apply fuzzy matching to population data's country names
population_df['Matched_Country'] = population_df['Country'].apply(fuzzy_match_country, args=(df_2024['Country'],))

# Merge population data with filtered existing DataFrame based on matched country names
merged_df = pd.merge(df_2024, population_df, how='left', left_on='Country', right_on='Matched_Country')

# Drop unnecessary columns
merged_df.drop(['Matched_Country'], axis=1, inplace=True)
merged_df['Population'] = merged_df['Population'].str.replace(',', '').astype(float)

# Drop rows where the population data is missing
merged_df.dropna(subset=['Population'], inplace=True)
pd.options.display.float_format = '{:.2f}'.format

# Save the merged DataFrame to a new CSV file
merged_df.to_csv("merged_data.csv", index=False)
merged_df.sample(4)
```

```python
def assign_emoji(score):
    if score >= 6 and score <= 8:
        return '<svg xmlns="http://www.w3.org/2000/svg" width="20" height="20"><circle cx="10" cy="10" r="10" fill="green"/><text x="50%" y="50%" text-anchor="middle" fill="white">ðŸ¤©</text></svg>'
    elif score >= 4.5 and score < 6:
        return '<svg xmlns="http://www.w3.org/2000/svg" width="20" height="20"><circle cx="10" cy="10" r="10" fill="yellow"/><text x="50%" y="50%" text-anchor="middle">ðŸ˜‘</text></svg>'
    elif score >= 1 and score < 4.5:
        return '<svg xmlns="http://www.w3.org/2000/svg" width="20" height="20"><circle cx="10" cy="10" r="10" fill="red"/><text x="50%" y="50%" text-anchor="middle" fill="white">ðŸ˜¡</text></svg>'
    else:
        return ''  # Return an empty string if no emoji matches
```

```python
df['Emoji'] = df.apply(lambda row: assign_emoji(row['Ladder score']), axis=1)
df.isna().sum()
```

```python
# Impute missing values for numerical columns with mean
numerical_cols = ['Economy', 'Social support', 'Health', 'Freedom', 'Trust', 'Generosity']
for col in numerical_cols:
    df[col].fillna(df[col].mean(), inplace=True)

# Impute missing values for Dystopia Residual with median
df['Dystopia Residual'].fillna(df['Dystopia Residual'].median(), inplace=True)

# Check if there are any missing values left
print(df.isnull().sum())

# Save the imputed dataset
df.to_csv("imputed_dataset.csv", index=False)
```

```python
import pandas as pd
from geopy.geocoders import Nominatim

# Read your dataset
data = pd.read_csv('imputed_dataset.csv')

# Create a geolocator object
geolocator = Nominatim(user_agent="my_geocoder")

# Function to get latitude and longitude for a country
def get_lat_lon(country):
    location = geolocator.geocode(country)
    if location:
        return location.latitude, location.longitude
    else:
        return None, None

# Add columns for latitude and longitude
data['Latitude'], data['Longitude'] = zip(*data['Country'].apply(get_lat_lon))

# Save the updated dataset
data.to_csv('imputed_dataset_with_coords.csv', index=False)

print("Latitude and longitude added successfully!")
```

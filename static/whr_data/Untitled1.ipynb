```python
df = pd.read_csv('imputed_dataset_with_coords.csv')
df.sample(5)
```

```python
pop1522=pd.read_csv("pop2015_2022.csv")
pop1522
```

```python
from fuzzywuzzy import process

# Function to find the closest match for a country name
def find_closest_match(country_name, population_data):
    matches = process.extract(country_name, population_data['Country'], limit=1)
    return matches[0][0] if matches and matches[0][1] > 70 else None


# Add population to the bigger table based on the year and country name
for index, row in df.iterrows():
    country_name = row['Country']
    year = row['Year']
    if pd.notnull(year):
        year = int(year)  # Convert to integer
        closest_match = find_closest_match(country_name, pop1522)
        if closest_match:
            population_col = str(year)
            if population_col in pop1522.columns:
                population = pop1522.loc[pop1522['Country'] == closest_match, population_col]
                if not population.empty:
                    df.at[index, 'Population'] = population.values[0]

df
```

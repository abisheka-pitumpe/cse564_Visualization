```python
import pandas as pd
from fuzzywuzzy import process
from fuzzywuzzy import fuzz

# Read the datasets
continents_df = pd.read_csv("continents.csv")

# Assuming your dataframe is already loaded in a variable called 'df'

def fuzzy_match_continent(row):
    # Get the country name from the current row
    country = row['Country']

    # Define a threshold for a good match (e.g., 80% similarity)
    threshold = 80

    # Extract the best match from 'continents_df' based on country name
    matches = process.extractOne(country, continents_df['Country'], scorer=fuzz.partial_ratio)

    # Check if a match is found and above the threshold
    if matches and matches[1] >= threshold:
        return matches[0]  # Return the matched country name
    else:
        return None  # Return None if no good match found

# Apply fuzzy matching to each row in 'df'
df['Matched_Country'] = df.apply(fuzzy_match_continent, axis=1)

# Merge DataFrames based on the matched country
df_with_continent = df.merge(continents_df[['Country', 'Continent']], how='left', left_on='Matched_Country', right_on='Country')

# Drop the temporary 'Matched_Country' column (optional)
df_with_continent = df_with_continent.drop('Matched_Country', axis=1)


# Display the resulting DataFrame
df = df_with_continent
```

```python
from fuzzywuzzy import process

# Read the longlat.csv file into a DataFrame
longlat_df = pd.read_csv("longlat.csv")

# Define a function for fuzzy matching
def fuzzy_match_country(name, choices):
    match = process.extractOne(name, choices)
    return match[0] if match[1] >= 80 else None

# Apply fuzzy matching to each country in df to find its closest match in longlat_df
df['Matched_Country'] = df['Country_x'].apply(lambda x: fuzzy_match_country(x, longlat_df['Country']))

# Merge the latitude and longitude columns from longlat_df to df based on the matched country names
df_merged = df.merge(longlat_df[['Country', 'Latitude', 'Longitude']], left_on='Matched_Country', right_on='Country', how='left')

# Drop the 'Matched_Country' column and the duplicate 'Country' column
df_merged = df_merged.drop(['Matched_Country', 'Country_y'], axis=1)

# Rename the 'Country_x' column to 'Country'
df_merged = df_merged.rename(columns={'Country_x': 'Country'})

# Display the resulting DataFrame
df_merged

```
